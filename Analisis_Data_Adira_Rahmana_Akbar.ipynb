**Merge sellers dan order items**

```python
sellersperformance_df = pd.merge(
    left=sellers_df,
    right=orderitems_df,
    how="right",
    left_on="seller_id",
    right_on="seller_id"
)
sellersperformance_df.head()
```

```python
performance_by_state = sellersperformance_df.groupby("seller_state", as_index=False).agg(
    total_orders=pd.NamedAgg(column="order_id", aggfunc="nunique"),
    total_revenue=pd.NamedAgg(column="price", aggfunc="sum")
).sort_values(by="total_orders", ascending=False)

print(performance_by_state)
```

```python
revenue_by_state = sellersperformance_df.groupby("seller_state", as_index=False).agg(
    total_orders=pd.NamedAgg(column="order_id", aggfunc="nunique"),
    total_revenue=pd.NamedAgg(column="price", aggfunc="sum")
).sort_values(by="total_revenue", ascending=False)

print(revenue_by_state)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

**Tipe Payment Terbanyak**

```python
grouped_data = orderpayments_df.groupby(by="payment_type").order_id.nunique().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data, color='skyblue')
plt.title('Distribution of Order Counts by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Number of Unique Order Items')
plt.xticks(rotation=45, ha='right')  # Rotasi label sumbu x untuk keterbacaan
plt.tight_layout()

plt.show()
```

### Pertanyaan 2:

**Performance Terbaik Penjual dalam Jumlah Pembelian di Negara Bagian (top 5)**



```python
top_order_states = performance_by_state.head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_order_states['seller_state'], top_order_states['total_orders'], color='skyblue')
plt.title('Top 5 States with the Highest Total Orders')
plt.xlabel('Seller State')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.gca().invert_xaxis()
plt.tight_layout()

plt.show()
```

**Performance Terburuk Penjual dalam Jumlah Pembelian di Negara Bagian (top 5)**

```python
bot_order_states = performance_by_state.tail(5)

plt.figure(figsize=(10, 6))
plt.bar(bot_order_states['seller_state'], bot_order_states['total_orders'], color='skyblue')
plt.title('Top 5 Bad Total Revenue by State')
plt.xlabel('Seller State')
plt.ylabel('Total Orders')
plt.xticks(rotation=45, ha='right')
plt.gca().invert_xaxis()
plt.tight_layout()

plt.show()
```

### Pertanyaan 3:

**Performance Terbaik Penjual dalam Jumlah Pendapatan di Negara Bagian (top 5)**

```python
top_revenue_states = revenue_by_state.head(5)

plt.figure(figsize=(10, 6))
plt.plot(top_revenue_states['seller_state'], top_revenue_states['total_revenue'], marker='o', color='skyblue', linestyle='-')
plt.xlabel('Negara Bagian')
plt.ylabel('Total Revenue')
plt.title('Top 5 Good Total Revenue by State (Line Chart)')

plt.gca().yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.2f}'))

plt.xticks(rotation=45)
plt.gca().invert_xaxis()
plt.grid(True)

plt.show()
```

**Performance Terburuk Penjual dalam Jumlah Pendapatan di Negara Bagian (top 5)**

```python
bot_revenue_states = revenue_by_state.tail(5)

plt.figure(figsize=(10, 6))
plt.plot(bot_revenue_states['seller_state'], bot_revenue_states['total_revenue'], marker='o', color='skyblue', linestyle='-')
plt.xlabel('Negara Bagian')
plt.ylabel('Total Revenue')
plt.title('Top 5 Bad Total Revenue by State (Line Chart)')

plt.gca().yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.2f}'))

plt.xticks(rotation=45)
plt.gca().invert_xaxis()
plt.grid(True)

plt.show()
```
